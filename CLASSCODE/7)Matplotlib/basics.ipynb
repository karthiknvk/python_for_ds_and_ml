```python
#OBJECT ORIENTED METHOD
```

```python
fig=plt.figure()

axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('SET TITLE')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.5,0.15,0.35,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title('AXES 1 TITLE')
axes2.set_title('AXES 2 TITLE')
```

```python
fig=plt.figure()
axes3=fig.add_axes([0.2,0.2,0.8,0.4])
axes3.plot(x,y)
```

#LEGENDS

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2)
ax.plot(x,x**3,)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X^2')
ax.plot(x,x**3,label='X^3')

ax.legend()
```

```python
fig.savefig('label_fig.png')
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X^2')
ax.plot(x,x**3,label='X^3')

ax.legend(loc=10)
```

#PLOT APPEARANCE

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color='purple',linewidth=5,alpha=0.5,linestyle='--')
#axes.plot(x,y,color='#FFFFFF',lw=5,linestyle='solid') can use RGB hex code,lw instead linewidth
```

#MARKERS

```python
x
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color='red',linewidth=5,marker='o',markersize=15,markerfacecolor='pink',markeredgecolor='black')
#marker marks x vs y points
```

#X LIMIT AND Y LIMIT

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',linewidth=2,marker='o',markersize=15,markerfacecolor='pink',markeredgecolor='black')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
