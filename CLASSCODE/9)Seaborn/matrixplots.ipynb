#HEATMAP

```python
tc=tips.corr(numeric_only=True)
tc
```

```python
sns.heatmap(tc,annot=True,cmap='hot')
```

```python
fc=flights.corr(numeric_only=True)
fc
```

```python
fp=flights.pivot_table(index='month',columns='year',values='passengers')
fp
```

```python
sns.heatmap(fp,)
```

```python
sns.heatmap(fp,cmap='magma',linecolor='black',linewidths=0.5)
#cmpa-> magma,coolwarm,hot,cool
```

#CLUSTER MAP

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```
