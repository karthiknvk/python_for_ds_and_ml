```python
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.countplot(data=tips,x='sex')
```

```python
len(tips[tips['sex']=='Male'])
```

```python
sns.set_style('ticks')
sns.countplot(data=tips,x='sex')
sns.despine()#remove top&right spines,also bottom and left(bottom=True,left=True)
```

```python
plt.figure(figsize=(12,3))
sns.countplot(data=tips,x='sex')
```
