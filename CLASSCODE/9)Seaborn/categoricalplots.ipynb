```python
import seaborn as sns
%matplotlib inline
tips=sns.load_dataset('tips')
tips
```

```python
tips.head()
```

#BARPLOT

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator='std')
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator='sum')
```

```python
tips['total_bill'].mean()
```

```python
tips[tips['sex']=='Male']['total_bill'].sum()
```

```python
tips[tips['sex']=='Female']['total_bill'].std()
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

#COUNTPLOT

```python
sns.countplot(x='sex',data=tips)
#same bar plot exept y not chosen, instead show occurrences(count) in  y axis
```

#BOXPLOT

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex',palette='cool',)
```

#VIOLINPLOT

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',palette='cool',split=True)
```

#STRIPPLOT

```python
sns.stripplot(x='day',y='total_bill',data=tips,palette='cool',hue='sex',)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False)
```

#SWARMPLOT

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,palette='cool')
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```
